```python


def world_cloud_auxiliar(category):
    # Sample text
    texts = train.loc[train['Category']==category]
    text = " ".join(texts['Text'])
    # Create a WordCloud object
    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=STOPWORDS).generate(text)
    
    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(category)
    plt.show()
```

### 3.- Embedding

We are going to use a pretrained sentence transformer to generate the sentence embeddings. 
This is done using a pretrained transformer model. refer to: https://huggingface.co/sentence-transformers.
For this exercise I'm picking a "small" model in the interest of time.

```python
%%time

def process(df):
    vectorizer = TfidfVectorizer(stop_words='english')  # Using stop words to exclude common words like "the", "and", etc.
    vectors = vectorizer.fit_transform(df['Text'])
    feature_names = vectorizer.get_feature_names_out()
    dense = vectors.todense()
    denselist = dense.tolist()
    new_df_train = pd.DataFrame(denselist, columns=feature_names)
    df_train = pd.concat([train, new_df_train], axis=1)
    return df_train, new_df_train

df_train, new_df_train = process(train)
```

```python
# we generate category mappings to numbers and the inverse mapping
map_categories = dict([(x[0],x[1]) for x in zip(categories,range(0,len(categories)))])
imap_categories = dict([(x[1],x[0]) for x in zip(categories,range(0,len(categories)))])
print(map_categories)
print(imap_categories)
```
